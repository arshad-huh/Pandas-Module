###`Q-8:` Team1 vs Team2 Dashbord
Create a function which will take two string(name of two teams) as input. Show win Loss record between them and player getting most player of the match award in matches between these two teams.
```
team1_vs_team2('Kolkata Knight Riders','Chennai Super Kings')
```

````python
def team_vs_team_stats(team1, team2):

    matches_between_teams = ipl[((ipl['Team1'] == team1) & (ipl['Team2'] == team2)) |
                                ((ipl['Team1'] == team2) & (ipl['Team2'] == team1))]

    win_loss_record = matches_between_teams['WinningTeam'].value_counts()

    most_player_of_the_match = matches_between_teams['Player_of_Match'].value_counts().idxmax()

    print(f"Win-Loss Record between {team1} and {team2}:\n{win_loss_record}")
    print(f"\nPlayer with most Player of the Match awards in matches between {team1} and {team2}: {most_player_of_the_match}")


team_vs_team_stats('Mumbai Indians', 'Chennai Super Kings')
````
